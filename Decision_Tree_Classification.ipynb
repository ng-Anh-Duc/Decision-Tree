```python
import pandas as pd
import sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
dataset = load_iris()
X = dataset.data
y = dataset.target
X[:5,:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
classifier = tree.DecisionTreeClassifier(criterion='gini', max_depth=4, min_samples_leaf=10)
classifier.fit(X_train, y_train)
fig, ax = plt.subplots(figsize=(10,10))
tree.plot_tree(classifier, ax=ax, feature_names=["sepal_length", "sepal_width", 'petal_length', 'petal_width'])
plt.show()
```

```python
from sklearn.metrics import accuracy_score

y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
classifier = tree.DecisionTreeClassifier(criterion='entropy', max_depth=4, min_samples_leaf=10)
classifier.fit(X_train, y_train)
fig, ax = plt.subplots(figsize=(10,10))
tree.plot_tree(classifier, ax=ax, feature_names=["sepal_length", "sepal_width", 'petal_length', 'petal_width'])
plt.show()
```

```python
from sklearn.metrics import accuracy_score

y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred)
```
